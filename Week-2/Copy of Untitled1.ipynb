```python
# Ask the user how many Fibonacci numbers to print
n_terms = int(input("Enter the number of terms: "))

# First two numbers
a, b = 0, 1

print("Fibonacci Sequence:")

for _ in range(n_terms):
    print(a, end=" ")
    a, b = b, a + b
```



*   Purpose: Prints the Fibonacci sequence where each term is the sum of the previous two, starting with 0 and 1.
*  Modification for Colab: Added input() so the number of terms can be entered interactively when running the cell.

*   Variables:
a → first term (0)
b → second term (1)
n_terms → number of terms to display (user input).
*  Loop: Uses a for loop to print terms in sequence, updating a and b at each iteration.
*  Output: Numbers printed on the same line, separated by spaces.
